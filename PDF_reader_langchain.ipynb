```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
from sentence_transformers import SentenceTransformer
```

```python
loader = PyPDFLoader("https://www.ibm.com/downloads/cas/7109RLQM")
pages = loader.load_and_split()
```

```python
print(len(pages))
```

```python
print(pages[1])
```

```python
'''
To use, you should have the sentence_transformers python package installed.
pip install sentence-transformers
'''

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
doc_embedding = hf.embed_query(pages[1].page_content)
```

```python
len(doc_embedding)
```

```python
type(pages[1].page_content)
```

```python
doc_embedding
```

```python
input_embedding = hf.embed_query("competitive advantage")
```

```python
len(input_embedding)
```

```python
input_embedding
```

```python
# all-mpnet-base-v2
# model = SentenceTransformer("all-MiniLM-L6-v2")
model = SentenceTransformer("all-mpnet-base-v2")
```

```python
similarities = model.similarity(doc_embedding, input_embedding)
```

```python
print(similarities)
```

```python
embeddings1 = model.encode([pages[1].page_content])
embeddings2 = model.encode(["competitive advantage"])
```

```python
print(len(embeddings1[0]),len(embeddings2[0]))
```

```python
similarities_mod_encode = model.similarity(embeddings1, embeddings2)
```
